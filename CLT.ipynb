```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import scipy.stats as stats
%matplotlib inline
```

```python
from scipy.stats import uniform

#setting the seed for reproducibility
np.random.seed(1)

#creating a uniform distribution of size 100000
uniform_pop=uniform.rvs(0,10,size=100000)

plt.hist(uniform_pop)
plt.title("Uniform Distribution Population")
plt.xlabel("X~U(0,10)")
plt.ylabel("Count")
plt.show()
```

### creating a sampling distribution from this population(sample_size=5,number of samples=500)
* drawing independent observations of sample size 5, n=5
* calculating the mean of these 5 observations
* repeating the above step for 500 times to calculate 500 sample means

```python
np.random.seed(1)

n=5

sample_means=[]
for i in range(500):
    sample=np.random.choice(uniform_pop,size=n)
    sample_mean=np.mean(sample)
    sample_means.append(sample_mean)

    
sns.displot(sample_means,kde=True)
plt.title("Distribution of sample means for n=" + str(n))
plt.xlabel("sample mean")
plt.ylabel("count")
plt.show()
```
